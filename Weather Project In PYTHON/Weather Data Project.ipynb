# Q.3-Find the no.of Times when the weather  is exactly clear

```python
# .value_counts()
wd["Weather"].value_counts()
```

```python
# filtering
# wdc=wd.loc[wd["Weather"].isin(["Clear"])]
# wdc
#wdc["Weather"].value_counts()


w=wd[wd.Weather=="Clear"]
w["Weather"].value_counts()
```
